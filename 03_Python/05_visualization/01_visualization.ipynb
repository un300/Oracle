##### 하나의 화면에 여러가지 그래프를 그릴때

 - subplot(2, 1, 1) : 2행 1열에 첫번째 그래프
 - subplot(2, 1, 2) : 2행 1열에 두번째 그래프

```python
# 그래프를 두개 그리기위해 의미없는 데이터를 만들어 줌

X1 = np.linspace(0.0, 5.0)
Y1 = np.cos(2 * np.pi * X1) * np.exp(-X1)
print(X1)
print(Y1)

X2 = np.linspace(0.0, 2.0)
Y2 = np.cos(2 * np.pi * X2)
print(X2)
print(Y2)
```

```python
# subplot을 그리는 방법1

axes01 = plt.subplot(1, 2, 1)
plt.plot(X1, Y1, 'rs--')
plt.title('subplot_01')

axes02 = plt.subplot(1, 2, 2)
plt.plot(X2, Y2, 'go-')
plt.title('subplot_02')

plt.tight_layout()  ## subplot간의 간격을 자동으로 맞추어 주는 함수
plt.show()
```

```python
# subplot을 그리는 방법2 : 객체의 형태로 subplot을 그리는 방법

axes01 = plt.subplot(1, 2, 1)
axes02 = plt.subplot(1, 2, 2)


axes01.plot(X1, Y1, 'rs--')
axes01.set_title('subplot01')   # set_title()함수 plt.title()의 차이점을 알아라


axes02.plot(X2, Y2, 'go--')
axes02.set_title('subplot02')

plt.tight_layout() 
plt.show()
```

```python
Y = [2, 3, 1]
X = np.arange(len(Y))


plot1 = plt.subplot(1, 2, 1)
plot2 = plt.subplot(1, 2, 2)

xlabel = ['1등석', '2등석', '3등석']

plot1.set_title('수직')
plot1.bar(X, Y)
plot1.set_xticks(X, xlabel)
plot1.set_yticks(sorted(Y))
plot1.set_xlabel('선실 등급')
plot1.set_ylabel('생존자 수')


plot2.set_title('수평')
plot2.barh(X, Y)
plot2.set_xticks(X, xlabel)
plot2.set_yticks(sorted(Y))
plot2.set_xlabel('선실 등급')
plot2.set_ylabel('생존자 수')


# subplot에서 조건으 넣을떈,
# [subplot객체].set_[함수]
# 이 형태로 적어주어야 하내!
```

## 실습

##### 문제01]
   - 자동차 배기량(displ)에 따라 고속도로연비(hwy)가 다른지 알아보자
   - 배기량이 4이하인 자동차와 5이상인 자동차 중 어떤 자동차의 고속도로연비가 평균적으로 더 높은가?

```python
f1 = lambda x : '4이하' if x <=4 else '4초과 5미만'
f2 = lambda y : '5이상' if y >=5 else f1(y)

data_df[ 'displ_class' ] = data_df[ 'displ' ].apply(f2)
data_df[ ['displ_class', 'hwy'] ].groupby('displ_class').agg(np.mean).plot.bar()
```

##### 문제 02] 
   - 자동차 제조 회사에 따라 도시 연비가 다른지 알아보려고 한다.
   - audi와 toyota 중 어느 manufacturer(자동차 제조회사)의 cty(도시연비)가 평균적으로 더 높은지 알아보시오

```python

index_audi = data_df['manufacturer'] == 'audi'
index_toyota = data_df['manufacturer'] == 'toyota'

want_df = data_df[ index_audi | index_toyota ] 
want_df[ ['manufacturer', 'hwy'] ].groupby('manufacturer').agg(np.mean).plot.barh()

## audi의 cty(도시연비)가 더 높다
```

##### 문제03]
   - chevorlet, ford, honda 자동차의 고속도로연비(hwy) 평균을 알아보려고 한다.
   - 이 회사들의 데이터를 추출한 후 hwy 전체 평균을 확인하라

```python
chevrolet_index   = data_df['manufacturer'] == 'chevrolet'
ford_index        = data_df['manufacturer'] == 'ford'
honda_index       = data_df['manufacturer'] == 'honda'

temp_df = data_df.loc[chevrolet_index | ford_index | honda_index, :]
mean_df = temp_df[ ['manufacturer', 'hwy'] ].groupby(temp_df.manufacturer).agg(np.mean)
mean_df.plot.bar()

display(mean_df)
print('세 자동차 회사의 hwy의 평균 : ', temp_df['hwy'].mean())
```

##### 문제04] 
   - hwy 변수의 결측값을 제외하고, 어떤 구동방식의 고속도로평균 연비가 높은지 알아보자.


```python
# 랜덤으로 hwy에 결측값 넣기
np.random.seed(123)
random_index = np.random.randint(0, hap_che_df.shape[0], 20)
hap_che_df.loc[random_index, ['hwy'] ] = np.NaN

# 구동방식(drv)에 따라 평균치 구하기
drop_na_df = hap_che_df.dropna()
drop_na_df.info()

# 구동방식별 연비평균을 비교하기 위해 막대그래프로 시각화
for_plot_df = drop_na_df.groupby('drv').mean()[['hwy']]

display(for_plot_df)
for_plot_df.plot.bar(rot=0)
plt.title('구동방식별 평균연비')
plt.xlabel('구동방식')
plt.ylabel('연비')
plt.grid(True)
```

##### 문제07]
 - 어떤 회사에서 생산한 suv 차종의 도시 연비가 높은지 알아보려고 한다.
 - suv 차종을 대상으로 평균 도시연비가 가장 높은 회사 다섯곳을 막대 바로 시각화 해보자.

```python
suv_df = car[car['class'] == 'suv']
pivot_df = suv_df.pivot_table('cty', 'manufacturer', 'class', aggfunc='mean')

pivot_df.columns = ['suv_average_cty']
big5_suv = pivot_df.sort_values(by='suv_average_cty', ascending=False).iloc[0:5, :]

big5_suv.plot.bar()
```
