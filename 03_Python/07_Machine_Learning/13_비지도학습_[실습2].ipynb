## 1. One-Hot Encoding 시행

##### 분석에 의미없는 칼럼 제거
 - `학교명`, `위도`, `경도`: OneHot인코딩의 의미가 없는 칼럼이다.
 - `주야` : 모두 주간으로 같기에 분석에 의미가 없는 칼럼

```python
# 학교명의경우 모두 다르기에 OneHot인코딩의 의미가 없다

encoding_df = df.drop(['학교명', '위도', '경도', '주야'], axis=1, inplace=False)
```

##### `지역`, `코드`, `유형`의 칼럼이 범주형이다
   - `지역`, `유형`은 문자형으로 되어있어 `get_dummies()`를 적용시키면 자동으로 OneHotEncoding을 수행한다
   - 하지만 `코드`의 경우는 수치형이기 때문에, 데이터타입을 문자형으로 바꾸어준다

```python
encoding_df['코드']  = encoding_df['코드'].map(lambda x :'코드_' + str(x))
```

##### `get_dummies`를 이용해 One Hot Encoding을 진행하자

```python
dummies_df = pd.get_dummies(encoding_df)
dummies_df.info()
```

## 2. PCA를 수행

##### 주성분의 수를 정해야한다.
   - 주성분의 수는 주성분의 개수에 따른 Variance(분산)이 차지하는 비율을 기준으로 하여 주관적으로 선택한다.
   - 주성분의 Variance가 크다는 것은 PCA를 수행한 Feature들에 대한 설명력이 높다는 주성분을 의미한다.

```python
pca = PCA(n_components = len(dummies_df.columns) - 1)

pca.fit(dummies_df)

Y = pca.explained_variance_[0:10]
X = np.arange(len(Y))[0:10]

sns.lineplot(x = X, y = Y, marker = 'o')
```

   - 이론적으로, 주성분이 2개일때 43개의 모든 Feature를 설명가능하다
   - 주성분을 1개로 하기는 좀 그러니.. **주성분을 2개로 PCA 를 진행하자**

```python
pca = PCA(n_components = 2)
pca.fit(dummies_df)

PCA_df = pd.DataFrame(pca.transform(dummies_df), columns=['PCA1', 'PCA2'])
```

## 3. K-means를 이용한 군집화
   - 군집화를 진행하기 전에 주성분 두개(PCA1, PCA2)를 각각 X, Y축으로 하여 그래프를 그려보자

```python
sns.scatterplot(x=PCA_df['PCA1'], y=PCA_df['PCA2'])
```

   - 뭔가.. 인위적인 그래프가 나타났다..
   - 또한, 딱히 군집으로 묶을만한 것처럼 보이지 않는다..
   - 그래서 군집의 개수에 따라 군집을 시각화 해주는 함수를 구현하여 가장 보기좋은 군집의 개수를 선택하여보자

```python
def cluster_visualization(df, n_cluster):
    kmeans = KMeans(n_clusters = n_cluster)
    kmeans.fit(df)
    df['target'] = kmeans.labels_
    
    sns.scatterplot(x=df['PCA1'], y=df['PCA2'], c=df['target'], cmap='Set2')
    
    return kmeans.labels_, df
```

```python
cluster_label, PCA_cluster_df = cluster_visualization(PCA_df, 5)
```

   - 나는 군집이 5개일때 제일 이뻐보여서 5개로 진행한다 ㅎㅎ..

## 4. 군집의 결과를 지도에 표시하자

##### 먼저 원본 데이터(`df`)에 군집화된 결과(`cluster_label`)를 입력하자

```python
df['cluster_label'] = cluster_label
```

##### 다음으로 리스트에 군집화의 결과가 같은 데이터프레임을 나누어 담자
   - 말이 어렵네요 이해가 안되면 아래 코드를 보고 이해해 주세요

```python
df_list = [df[df['cluster_label']==i] for i in range(len(set(cluster_label)))]
```

##### 이제 지도에 군집화 결과를 시각화 해보자
   - 제공받은 data의 Feature들에 근거하여 Kmeans가 비슷하다고 생각하는 학교는 같은 색으로 지도에 표시될 것이다!
   - 연습과 실습을 위해서 진행하였기에... 분석결과가 정확하다고 할 수 없습니다. 오직 연습용입니다

```python
school_map = folium.Map(location=[37.55, 126.98], titles='Stamen Terrain',
                       zoom_start=12)


k=0
color_list = ['red', 'blue', 'black', 'green', 'skyblue']
for temp_df in df_list:
    for name, latitude, longtitude in zip(temp_df.학교명, temp_df.위도, temp_df.경도):
        folium.CircleMarker(
            [latitude, longtitude],
            radius=5,
            color=color_list[k],
            fill = True,
            fill_color = 'black',
            fill_opacity = 0.5,
            popup = name).add_to(school_map)
    k+=1

school_map
```
