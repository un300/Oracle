## 1. 크롤러란?
> 크롤링을 하는 프로그램을 크롤러라 한다.

## 2 .urllib 패키지
> - 인터넷에서 데이터를 받아 파이썬으로 불러오는 기능들이 들어있는 패키지이다. `urllib`는 크롤링 라이브러리이다.
> - `urllib`는 파이썬에 기본적으로 내장되어 있는 라이브러리이기 때문에 바로 import하여 사용할 수 있다.

## 3. 네이버에서 첫 페이지 받아오기
> 1. `www.naver.com` 페이지에서 마우스오른쪽을 클릭하여 페이지 소스보기를 클릭한다.
> 2. 그 다음 아래와 같이 코드를 작성한다.

```python
from urllib.request import urlopen

url = "https://www.naver.com/"
html = urlopen(url)

print(html.read())
```

> 위의 코드를 작성하면, 콘솔창에 어마무시한 한줄이 길게 나타날 것이다. 이 코드는 아까 `페이지 소스보기`에서 본 html코드와 동일하다.

###### 여기서 `크롤링`이라 함은 홈페이지로 부터 위의 코드를 가져오는 것을 뜻한다.
###### 코드를 가져와서 필요한 정보를 추출해내는 과정은 `파싱`이라고 한다.
###### `크롤링`을 수행하는 라이브러리는 `urllib`이고,
###### `파싱`을 수행하는 라이브러리는 `BeautifulSoup`이다

## 4. 뷰티풀솝(BeautifulSoup) 설치하기

> 뷰티플솝은 외부 라이브러리이기 때문에 파이참의 경우 따로 설치해 주어야한다. 하지만 우리의 경우는 쥬피터 노트북이기 때문에 바로 import하여 사용할 수 있다

> 뷰티풀솝을 임포트한 후에, 다음과 같은 예시코드를 작성해보자

```python
import bs4

html_str = "<html><div>hello</div></html>"
bs_obj = bs4.BeautifulSoup(html_str, "html.parser")

print(type(bs_obj))
print(bs_obj)
print(bs_obj.find("div"))
```


#### bs4.BeautifulSoup(html_str, "html.parser")
> `html_str`이라는 html문서를 `pasing(파싱)` 하기위한 밑작업이라고 생각하면된다.



`print(type(bs_obj))` : bs_obj의 타입을 출력해줍니다

`print(bs_obj)` : bs_obj의 값을 출력합니다

`print(bs_obj.find("div"))` : div 속성의 값을 출력해 줍니다.


##### 좀더 html 스러운 예제를 통해 `hello`라는 단어만을 추출해 봅시다

```python
import bs4

html_str = """
<html>
    <body>
        <ul>
            <li>hello</li>
            <li>bye</li>
            <li>welcome</li>
        </ul>
    </body>
</html>
"""

bs_obj = bs4.BeautifulSoup(html_str, "html.parser")

ul = bs_obj.find("ul")
li = ul.find("li")
print(li)

```

`<li>`와 같은 테그를 없에고 싶으면 아래와 같이 `.text`를 사용하자.

```python
import bs4

html_str = """
<html>
    <body>
        <ul>
            <li>hello</li>
            <li>bye</li>
            <li>welcome</li>
        </ul>
    </body>
</html>
"""

bs_obj = bs4.BeautifulSoup(html_str, "html.parser")

ul = bs_obj.find("ul")
li = ul.find("li")
print(li.text)
```

## 5. `.findAll()` 사용하기

> - `.findall()`은 조건에 해당하는 모든 요소들을 리스트[] 형태로 추출해주는 기능이다.
> - `find()`는 만나는 첫 번째 태그만을 리턴해주는 함수이지만, `.findall()`은 조건에 해당하는 모든 요소를 리스트[]로 리턴해준다.

그렇다면 위의 `html_str`의 예제를 통해 `bye`를 추출해보자

```python
import bs4

html_str = """
<html>
    <body>
        <ul>
            <li>hello</li>
            <li>bye</li>
            <li>welcome</li>
        </ul>
    </body>
</html>
"""


bs_obj = bs4.BeautifulSoup(html_str, "html.parser")

ul = bs_obj.find("ul")
li_list = ul.findAll("li")
print(li_list)
print(li_list[1].text)
```

## 6. 테그와 속성 그리고 속성값

> - 뷰티풀솝을 사용하려면 tag(테그)와 property(속성)에 대해 알아야 한다. 또한, 뷰티풀솝을 사용하려면 HTML에대한 지식이 있어야한다.

#### 1) HTML
- HTML은 `<html>`로 시작해서 `</html>`로 끝나는 문서이다. HTML은 테그와 속성으로 구성되어있다.

#### 2) HTML - tag(테그)
- 테그에 대해 간략히 알아보면, `<ul></ul>`, `<li></li>`, `<div></div>`, `<a></a>` 등 `<>`로 감싸져 있는 것을 테그라고한다.    

#### 3) HTML - property(속성)
- `<ul class="greet">` 여기에서 `class`가 속성이다. (`=greet`는 속성이 아니라 "속성값"이라고 한다! "속성"과 "속성값" 이 둘을 잘 구분할 필요가 있다.)





###### 아래의 `html_str`에서 테그, 속성, find()만을 사용하여  두번째 `<ul class="reply">`를 추출해보라 (findAll()은 사용하지 말고)

```python
html_str = """
<html>
    <body>
        <ul class="greet">
            <li>hello</li>
            <li>bye</li>
            <li>welcome</li>
        </ul>
        <ul class="reply">
            <li>ok</li>
            <li>no</li>
            <li>sure</li>
        </ul>
    </body>
</html>
"""

bs_obj = bs4.BeautifulSoup(html_str, "html.parser")
ul = bs_obj.find("ul", {"class" : "reply"})
print(ul) 

```

## 7. 속성값 뽑아내기
> - a 테그에서 링크를 뽑아낼 때는 href라는 속석의 속성값을 뽑아내야 한다. 다음 `html_url`이라는 HTML에서 네이버 주소를 뽑아내보자
> - `<a href="https://www.naver.com">네이버</a>` 이 분에서 `www.naver.com`은 속성값이다. 그러므로 속성값을 추출하는 방법을 알아야한다.

```python
import bs4

html_url = """
<html>
    <body>
        <ul class="ko">
            <li>
                <a href="https://www.naver.com">네이버</a>
            </li>
            <li>
                <a href="https://www.daum.com">다음</a>
            </li>
        </ul>
        <ul class="sns">
            <li>
                <a href="https://www.google.com">구글</a>
            </li>
            <li>
                <a href="https://www.facebook.com">페이스북</a>
            </li>
        </ul>
    </body>
</html>
"""

bs_obj = bs4.BeautifulSoup(html_url, "html.parser")
atag = bs_obj.find("a")
print(atag)
print(atag['href'])

```

위와 같이 속성값을 추출하기 위해서는 딕셔너리 형태와 같이 추출하고자 하는 속성이름을 키값으로 여겨 추출하면 된다.\

위의 예제는 `href`라는 속성의 속성값을 추출해야하기에 `atag['href']`라고 지정하여 추출함을 볼 수 있다.

## 8. 네이버에서 특정 글자 추출하기
> - 처음부터 어려운 난이도의 프로그램을 만들면 시작부터 지치기 때문에, 비교적 쉬운 특정위치에 있는 단어 두개만 뽑아보자
> - 우리의 경우 네이버 맨 오른쪽 위에있는 `네이버를 시작페이지로`를 추출해보자

#### 1) urllib로 네이버 첫 페이지 데이터 받아오기

    1. 먼저 `urllib`패키지를 이용하여 `크롤링`을 수행한다

```python
import urllib.request

url = "https://www.naver.com"
html = urllib.request.urlopen(url)

print(html.read())
```

    2. 다음으로 `파싱`을 수행하기위해 BeautifulSoup 패키지를 사용한다

```python
import bs4
import urllib.request

url = "https://www.naver.com"
html = urllib.request.urlopen(url)

bs_obj = bs4.BeautifulSoup(html, "html.parser")

print(type(bs_obj))
print(bs_obj)
```

    3. '네이버를 시작페이지로'가 포함되어있는 div를 찾아 추출해보자
        크롬을 이용해 네이버에 접속한 후, F12키를 눌러 '네이버를 시작페이지로'의 코드 작성 위치를 찾자
   

```python
import bs4
import urllib.request

url = "https://www.naver.com"
html = urllib.request.urlopen(url)

bs_obj = bs4.BeautifulSoup(html, "html.parser")

start_page_div = bs_obj.find("div", {"class" : "service_area"})
print(start_page_div)
```

     4. 이제 위에서 실행한 코드를 바탕으로 '네이버를 시작페이지로'라는 문자를 추출해보자

```python
import bs4
import urllib.request

url = "https://www.naver.com"
html = urllib.request.urlopen(url)

bs_obj = bs4.BeautifulSoup(html, "html.parser")

start_page_div = bs_obj.find("div", {"class" : "service_area"})

final_target = start_page_div.find("a", {"class" : "link_set"})
print(final_target)
print()
print(final_target.text)
```
