```python
import warnings
warnings.filterwarnings('ignore')

from selenium import webdriver
from selenium.webdriver.support.select import Select
from tqdm import tqdm_notebook

import time
import pandas as pd
import numpy as np
```

## 1. 네이버쇼핑 : 에이제이월드 몰 접속

```python
path='../chromedriver/chromedriver.exe'
driver = webdriver.Chrome(path)
url = 'https://smartstore.naver.com/ajwmall'
driver.get(url)

time.sleep(3)
```

## 2. 에이제이월드 몰에 있는 모든 상품과 후기 크롤링

```python
# 1. 상품정보 리스트
total_price          = []
total_review_count   = []
total_review_point   = []
total_product_number = []

# 2. 개별후기댓글 리스트
indivisual_review_list  = []
indivisual_comment_list = []
product_no_list         = []

for i in range(3):
    url = "https://smartstore.naver.com/ajwmall/category/12217026edff49f18308dc0af270fda0?st=POPULAR&free=false&dt=LIST&page=" + str(i+1) + "&size=40"
    driver.get(url)
    time.sleep(3)
    ul_list = driver.find_elements_by_css_selector('ul._2XA8xyEsOb > li')

    print('현재페이지 : ', i+1)
    
    for idx in tqdm_notebook(range(len(ul_list))):
        ul_list[idx].click()

        time.sleep(2)

        # 1. 상품정보 크롤링
        price           = driver.find_element_by_css_selector('strong.aICRqgP9zw > span._1LY7DqCnwR').text
        review_content  = driver.find_elements_by_css_selector('div.NFNlCQC2mv > div._2Q0vrZJNK1')
        review_count    = driver.find_element_by_css_selector('span.q9fRhG-eTG').text
        review_point    = review_content[1].find_element_by_css_selector('strong').text
        product_number  = driver.find_element_by_css_selector('td.ABROiEshTD').text

        total_price.append(price)
        total_review_count.append(review_count)
        total_review_point.append(review_point)
        total_product_number.append(product_number)


        # 2. 상품평 크롤링
        # 상품평 크롤링 시 페이지 이동을 위한 리스트 만들기
        temp_page_list = driver.find_elements_by_css_selector('div._3YOQJDR2-0 > div._1HJarNZHiI > a')
        page_list      = temp_page_list[1:len(temp_page_list)-1]

        for page in page_list:
            page.click()
            time.sleep(2)
            li_list = driver.find_elements_by_css_selector('ul.TsOLil1PRz > li')
            for li in li_list:
                indivisual_review  = li.find_element_by_css_selector('div._2V6vMO_iLm > em').text
                indivisual_comment = li.find_element_by_css_selector('div.YEtwtZFLDz > span._3QDEeS6NLn').text

                indivisual_review_list.append(indivisual_review)
                indivisual_comment_list.append(indivisual_comment)
                product_no_list.append(product_number)

        # 뒤로가기
        driver.back()
        time.sleep(2)
        ul_list = driver.find_elements_by_css_selector('ul._2XA8xyEsOb > li')
        
        print('상품크롤링수     : ', len(total_price))
        print('상품후기크롤링수 : ', len(product_no_list))
        
```

## 3. 크롤링한 데이터 정제

```python
data = {
    '제품번호'      : total_product_number,
    '가격'          : total_price,
    '리뷰수'        : total_review_count,
    '평균리뷰점수'  : total_review_point
}
product_info_df = pd.DataFrame(data)
```

```python
product_info_df
```

```python
product_info_df['가격']         = product_info_df['가격'].str.replace('[^0-9]', '').astype(int)
product_info_df['평균리뷰점수'] = product_info_df['평균리뷰점수'].map(lambda x : x[0:3])
```

```python
product_info_df
```

```python
data = {
    '제품번호'      : product_no_list,
    '개별리뷰점수'  : indivisual_review_list,
    '리뷰내용'      : indivisual_comment_list
}
product_review_comment_df = pd.DataFrame(data)
```

```python
product_review_comment_df
```

```python
product_review_comment_df['리뷰내용'] = product_review_comment_df['리뷰내용'].str.replace('\n', ' ')
```

## 4. CSV파일로 저장

```python
product_info_df.to_csv('product_info.csv', encoding='UTF-8-SIG')
product_review_comment_df.to_csv('product_review.csv', encoding='UTF-8-SIG')
```
