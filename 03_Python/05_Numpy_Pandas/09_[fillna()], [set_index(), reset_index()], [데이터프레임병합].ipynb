###### 타이타닉 데이터를 이용한 fillna() 실습

```python
import seaborn as sns
titanic = sns.load_dataset('titanic', engine='python')
```

```python
# 실습

# 타이타닉 승객 중 나이를 명시하지 않은 고객은 나이를 명시한 고객의 평균 나이 값으로 대체

titanic['age'] = titanic['age'].fillna( titanic['age'].mean() )
print('타이타닉 df의 "Age"칼럼 결측치의 수 : ',  pd.isnull(titanic['age']).sum() )
display(titanic['age'].head(20))

# 타이타닉 승객에 대해 나이와 성별에 의한 age_gender_cat 열을 만들고
# 조건1 : 성별을 나타내는 문자열 male 또는 female로 시작한다.
# 조건2 : 성별을 나타내는 문자열 뒤에 나이를 나타내는 문자열로 변경하라
# 예시) 남성의 나이가 27이라면 > male27

titanic['age_gender_cat'] = titanic['sex'] + titanic['age'].astype(int).apply(str)
display(titanic['age_gender_cat'].head(20))
```

### 2. 데이터프레임 인덱스 조작하는 방법
   - set_index : 기존 행 인덱스를 제거하고 데이터프레임 내에 기존에 존재하는 열 중 하나를 인덱스로 설정
   - reset_index : 기존 행 인덱스를 데이터 프레임의 열로 추가한 후, 새로운 인덱스를 부여하는 함수

```python
np.random.seed(100)
index_df = pd.DataFrame(np.vstack([list('ABCDE'),
                                  np.round(np.random.rand(3, 5), 2)]).T,
                       columns=['col01', 'col02', 'col03', 'col04'])
index_df
```

```python
index_df2 = index_df.set_index('col01')
display(index_df)
display(index_df2)

# 아래의 결과를 보면 기존의 index가 사라지고 기존의 열을 index로 여김을 알 수 있다.
```

```python
index_df3 = index_df2.set_index('col02')
display(index_df2)
display(index_df3)

# 아래의 결과를 보면 기존의 index가 사라지고 기존의 열을 index로 여김을 알 수 있다.
# 확실히 이해 되죠잉?
```

```python
reset_index_df = index_df2.reset_index()
drop_reset_index_df = index_df2.reset_index(drop=True)

display(index_df2)
display(reset_index_df)
display(drop_reset_index_df)

## 각각의 결과를 비교해서 내것으로 만드세요
## 별로 어렵지 않아요
```

##### set_index() / reset_index() 실습

```python
# 5명의 학생의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 만든다.

# 1. 
# 학생 이름을 나타내는 열을 포함시키지 않고 
# 데이터프레임 df_score1 을 생성한 후, 
# df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다. 
# reset_index 명령으로 이 인덱스 열을 명령으로 
# 일반 데이터열로 바꾸여 데이터프레임 df_score2을 만든다.

data = {
    '국어' : [50, 65, 87, 10, 100, 89, 32, 83, 100, 58],
    '영어' : [100, 30, 20, 11, 88, 97, 100, 100, 10, 57],
    '수학' : [100, 59, 88, 29, 99, 29, 22, 99, 29, 50]
}

df_score1 = pd.DataFrame(data)
df_score1.index = ['철수', '영희', '민수', '짱구', '지민', '광수', '맹구', '지수', '수민', '설희']
df_score1.index.name = '이름'

df_score2 = df_score1.reset_index()

display( df_score1 )
display( df_score2 )

# 2.
# 학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 df_score2에 
# set_index 명령을 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다.
df_score3 = df_score2.set_index('이름')
display( df_score3 )
```

### 3. DataFrame merge

   - pd.merge( df1, df2, on='공통의칼럼', how='left'/'right' )

```python
data1 = {
    '학번' : [1, 2, 3, 4],
    '이름' : ['섭섭해', '김한준', '김선림', '최호진'],
    '학년' : [2, 4, 1, 3]
}

data2 = {
    '학번' : [1, 2, 4, 5],
    '전공' : ['컴퓨터사이언스', '수학', '수학', '컴퓨터사이언스'],
    '학점' : [3.3, 3.7, 4.5, 2.5]
}
```

```python
student_df = pd.DataFrame(data1)
major_df = pd.DataFrame(data2)

display(student_df)
display(major_df)
```

```python
pd.merge(student_df, major_df, on='학번', how='left')
```

```python
pd.merge(student_df, major_df, on='학번', how='right')
```

```python
pd.merge(student_df, major_df, on='학번', how='outer')
```

##### 공통의 컬럼이 없는 두 데이터프레임을 merge시킬때,
   - left_on / right_on 옵션을 부여하여 pd.merge() 함수를 사용한다. 

```python
data1 = {
    '학번' : [1, 2, 3, 4],
    '이름' : ['섭섭해', '김한준', '김선림', '최호진'],
    '학년' : [2, 4, 1, 3]
}

data2 = {
    '과목코드' : [1, 2, 4, 5],
    '전공' : ['컴퓨터사이언스', '수학', '수학', '컴퓨터사이언스'],
    '학점' : [3.3, 3.7, 4.5, 2.5]
}
```

```python
student_df = pd.DataFrame(data1)
major_df = pd.DataFrame(data2)

display(student_df)
display(major_df)
```

```python
pd.merge(student_df, major_df, left_on='학번', right_on='과목코드', how='inner')
```

##### 칼럼 인덱스가 아닌, 인덱스를 기준으로 병합한다면?
    - right_index=True

```python
pop_df01 = pd.DataFrame({
    'city' : ['seoul', 'seoul', 'seoul', 'busan', 'busan'],
    'year' : [2010, 2005, 2020, 2018, 2015],
    'pop'  : [1234567, 2345678, 3456789, 4567890, 5678901]
})

pop_df01
```

```python
pop_df01 = pd.DataFrame({
    'city' : ['seoul', 'seoul', 'seoul', 'busan', 'busan'],
    'year' : [2010, 2005, 2020, 2018, 2015],
    'pop'  : [1234567, 2345678, 3456789, 4567890, 5678901]
})

display(pop_df01)


pop_df02 = pd.DataFrame(np.arange(12).reshape( (6, 2) ),
    index   = [
                  ['busan', 'busan', 'seoul', 'seoul', 'seoul', 'seoul'],
                  [2010, 2005, 2020, 2018, 2015, 2010]
              ],
    columns = ['col01', 'col02']
)

display(pop_df02)
```

```python
pd.merge(pop_df01, pop_df02, left_on=['city', 'year'], right_index=True)
```

 - 다중 인덱스가 아닌 단일 인덱스를 이용하는 방법

```python
data1 = { "이름" : ["이지안","박동훈","이순신","강감찬"],
          "학년" : [2,4,1,3]}

data2 = { "학과" : ["CS","MATH","MATH","CS"],
          "학점" : [3.4,2.9,4.5,1.2]}
```

```python
df01 = pd.DataFrame(data1, index=[1, 2, 3, 4])
df02 = pd.DataFrame(data2, index=[1, 2, 4, 5])

display(df01)
display(df02)
```

```python
merge_df = pd.merge(df01, df02, left_index=True, right_index=True)
display(merge_df)
```

##### merge와 같은 함수  [데이터프레임1].join([데이터프레임2])

```python
temp_df01 = df01.join(df02, how='inner')
temp_df02 = pd.merge(df01, df02, left_index=True, right_index=True)

display(temp_df01)
display(temp_df02)
```
