## 1. 지도 시각화를 위한 라이브러리 : `folium`
   - '멀티캠퍼스'의 위치를 `folium`라이브러리를 활용해 지도에 표시해보자!

```python
import folium as g
```

```python
# 지도에 원하는 위치 보이게하기
g_map = g.Map(location=[37.509671, 127.055517], 
              zoom_start=18,
             tiles='Stamen Terrain')
g_map

# 지도에 원하는 위치 찍기
marker01 = g.Marker([37.509671, 127.055517],
                 popup='Campus Seven',
                 icon = g.Icon(color='blue'))
marker01.add_to(g_map)


# 지도에 원하는 반경(원) 나타내기
marker02 = g.CircleMarker([37.509671, 127.055517],
                         radius=100,
                         color='skyblue',
                         popup='Campus Seven',
                         fill_color='skyblue')
marker02.add_to(g_map)


# 지도 저장하기
g_map.save('./folium결과파일/now_location_map.html')

# 지도 실행하기
g_map
```

## 2. '서울지역 대학교 위치' 데이터

```python
# 파일 읽기 '서울지역 대학교 위치'

location_data = pd.read_excel('./실습데이터/서울지역 대학교 위치.xlsx')
```

```python
location_data.head()
location_data.columns = ['대학교', '위도', '경도']
```

```python
seoul_map01 = g.Map(location=[37.55, 126.98],
                 tiles='Stamen Terrain',
                 zoom_start=12)

# 대학교 위치 지도에 찍기
for idx in range(location_data.shape[0]):
    marker = g.Marker(location_data.loc[idx, ['위도', '경도']],
                       popup=location_data.loc[idx, '대학교'],
                       icon = g.Icon(color='blue'))
    marker.add_to(seoul_map01)
    
seoul_map01
```

```python
seoul_map02 = g.Map(location=[37.55, 126.98],
                 tiles='Stamen Terrain',
                 zoom_start=12)

# 대학교 위치 지도에 찍기
# 강사님 방법 : zip함수의 사용법을 다시한번 인지하게됨..
for name, latitude, longtitude in zip(location_data.대학교, location_data.위도, location_data.경도):
    marker= g.CircleMarker([latitude, longtitude],
                       popup = name,
                       icon  = g.Icon(color='red'),
                       radius=10,
                       fill=True,
                       fill_color='coral',
                       fill_opacity=0.3)
    marker.add_to(seoul_map02)
                
    
seoul_map02
```

## 4. '경기도 인구' 데이터

```python
# 데이터불러오기
pop_df = pd.read_excel('./실습데이터/경기도인구데이터.xlsx', index_col='구분')
pop_df.columns = pop_df.columns.map(str)
pop_df.head(10)
```

```python
display(pop_df.columns)
display(pop_df.index)
```

```python
# json 불러오기
import json

geo_data = json.load( open('./실습데이터/경기도행정구역경계.json', encoding='UTF-8') )
```

```python
# 지도생성
gyeonggi_map = g.Map(location=[37.55, 126.98],
                 tiles='Stamen Terrain',
                 zoom_start=9)


# 구분된 경계를 표시할때 사용하는 객체
g.Choropleth(geo_data = geo_data,
             data     = pop_df['2017'],
             columns  = [pop_df.index, pop_df['2017']],
             key_on   = 'feature.properties.name').add_to(gyeonggi_map)

gyeonggi_map
```
