 ## 2. [실습] year2020_baby_name.csv

```python
name_df = pd.read_csv('./실습데이터/year2020_baby_name.csv')
```

```python
name_df
```

 - 정렬     : sort_values(by=, ascending=)
 - 타입변환 : astype(type)

```python
sort_df = name_df.sort_values(by='COUNT', ascending=False)
sort_df
```

```python
rank_df = name_df.sort_values(by='COUNT', ascending=False)['COUNT'].rank(ascending=False)
rank = rank_df.astype('int64')
sort_df['RANK'] = rank
sort_df
```

```python
# gender를 기준으로 남자(M)데이터 프레임, 여자(F)데이터 프레임을 만들어라
male_df = sort_df[ sort_df['GENDER'] == 'M' ]
female_df = sort_df[ sort_df['GENDER'] == 'F' ]
```

```python
display(male_df.head())
display(female_df.head())

## 아래 결과를 보면 `male_df`와 `female_df`의 index가 다른것을 확인할 수 있다.
```

```python
## 새롭게 만들어진 데이터프레임의 인덱스를 초기화 하는 방법
male_df = male_df.reset_index(drop=True)
male_df.head()
```

```python
# pd.merge : 행 index가 같은 행 끼리 데이터를 가로뱡향으로 합친다
# 의미없는 데이터프레임이 만들어지지만, 한번 만들어보자
gender_df = pd.merge(male_df, female_df, left_index=True, right_index=True)
gender_df.head()


# COUNT_x에 대해서 정렬을 해보면?
rank_COUNT_x = gender_df.sort_values(by='COUNT_x', ascending=False)['COUNT_x'].rank(ascending=False).astype('int64')
gender_df['rank_COUNT_x'] = rank_COUNT_x
gender_df
```
