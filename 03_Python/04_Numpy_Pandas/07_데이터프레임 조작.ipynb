```python
np.random.seed(123)
count_df = pd.DataFrame(np.random.randint(5, size=(4, 4)), dtype=np.float)
display(count_df)
```

```python
# 임의로 NaN값을 한번 넣어보자
random_col = np.random.choice(3, replace=False, size=2)
random_row = np.random.choice(3, replace=False, size=2)

for col in random_col :
    for row in random_row :
        count_df.iloc[row, col] = np.NaN
display(count_df)
display(count_df.count(axis= 0))
```

##### 정렬
   - sort_index   : 열 인덱스 이름이나 행 인덱스 이름을 기준으로 정렬해준다
   - sort_values  : 칼럼이 가진 값을 기준으로 정렬해준다.

```python
np.random.seed(100)
sort_df = pd.DataFrame(np.random.randint(0, 10, (6, 4)))
display(sort_df)
```

```python
sort_df.columns = ['A', 'B', 'C', 'D']
sort_df.index   = pd.date_range('20201014', periods=6) # pandas에는 이런 함수도 있다
display(sort_df)
```

```python
# 행 뒤섞기 : 순열 랜덤 치환
random_date = np.random.permutation(sort_df.index)
random_date
sort_df2 = sort_df.reindex(index = random_date)    # df.reset_index()와는 다른 함수이다.
sort_df2

# 열 뒤섞기
sort_df3 = sort_df2.reindex(index=random_date,
                           columns=['B', 'A', 'D', 'C'])
sort_df3
```

```python
# 열 인덱스 이름을 기준으로 정렬
display( sort_df3.sort_index(axis=1) )

# 행 인덱스 이름을 기준으로 정렬
display( sort_df3.sort_index(axis=0) )
```

```python
# sort_values()도 한번 더 복습해보자 ㅎㅎ
display( sort_df3.sort_values(by='A', ascending=False) )
display( sort_df3.sort_values(by=[ 'A', 'B' ], ascending=False) )
```

##### 행렬의 합을 구할때 sum(axis=0/1)

```python
# 행방향 합
sort_df3['row_sum'] = sort_df3.sum(axis=1)
display( sort_df3 )

# 열방향 합
sort_df3.loc['col_sum', :] = sort_df3.sum(axis=0)
display( sort_df3 )
```

##### apply변환
   - 행이나 열 단위로 복잡한 데이터 가공이 필요한 경우 사용하는 함수이다.
   - lambda 식
   - apply는 함수를 인자로 넘겨 받을 수 있다.

```python
def get_square(a):
    return a**2
print("제곱근 :", get_square(3))
```

```python
# 위 코드를 lambda식으로 바꾼다면?
lambda_squre = lambda x : x**2   
print('제곱근 :', lambda_squre(3))
```

```python
np.random.seed(100)
apply_df = pd.DataFrame(np.random.randint(0, 10, (6, 4)))
apply_df.columns = ['A', 'B', 'C', 'D']
apply_df.index = pd.date_range('20201014', periods=6)
apply_df
```

```python
# 각 행의 column에 대해서 최대값 - 최소값을 구하여 새로운 column을 추가하고 싶다.
# 각 column안에서 최대값 - 최소값을 구해 출력
func = lambda x : x.max() - x.min()

# 행에 대한 apply
apply_df['row별 최대-최소'] = apply_df.apply(func, axis=1)
display(apply_df)

# 열에 대한 apply
apply_df.loc['col별 최대-최소', : ] = apply_df.apply(func, axis=0)
display(apply_df)
```

```python
# 실습
# 어려울겁니다

import seaborn as sns
titanic = sns.load_dataset('titanic')

### 1. embark_town의 문자열 개수를 별도의 칼럼인 embark_len 칼럼을 추가
# embark_town이 결측치인 경우 길이가 0인 값 대입
null_index = pd.isnull(titanic['embark_town'])
titanic.loc[null_index, 'embark_town'] = ''

# embark_town의 문자 길이를 구하고 embark_len 칼럼만들기
func = lambda x : len(x)
titanic['embark_len'] = titanic['embark_town'].apply(func)


### 2. if ~ else : 절을 활용하여 나이가 15세 이하면 child 그렇지 않으면 adult로 구분하는 child_adult 칼럼을 추가하라
func = lambda x : 'child' if x <= 15 else 'adult'
titanic['child_adult'] = titanic['age'].apply(func)

# age가 결측치인 경우 따로 처리
titanic.loc[ pd.isnull(titanic['age']), 'child_adult'] = '알수없음'


### 확인하기
display( titanic.head(10) )
```
