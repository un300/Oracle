 ## 2. df.iloc[행순서, 열순서]

```python
sample_df = pd.DataFrame(np.arange(10, 22).reshape(3, 4),
                         index   = ['a', 'b', 'c'],
                         columns = ['A', 'B', 'C', 'D'])
sample_df
```

```python
display(sample_df)
display(sample_df.iloc[0, 2:4].values)
display(sample_df.iloc[0, -2:].values)
display(sample_df.iloc[2, 1:3].values)
```

###### 데이터 갯수를 세알려보자
   - count()
       > - axis=0 : 가로방향으로 개수 세어줌
       > - axis=1 : 세로방향으로 개수 세어줌

```python
s = pd.Series(range(10))
display(s)
display(s.count())
```

```python
# count() 함수는 NaN값 (결측치)를 세지 않는다.
s[5] = np.NaN
s[2] = np.NaN
display(s.count())
```

```python
np.random.seed(123)
count_df = pd.DataFrame(np.random.randint(5, size=(4, 4)), dtype=np.float)
display(count_df)
```

```python
# 임의로 NaN값을 한번 넣어보자
random_col = np.random.choice(3, replace=False, size=2)
random_row = np.random.choice(3, replace=False, size=2)

for col in random_col :
    for row in random_row :
        count_df.iloc[row, col] = np.NaN
display(count_df)
display(count_df.count(axis= 0))
```

##### df['칼럼이름'].value_counts()
   - 하나의 칼럼에서 각 범주의 개수를 세어줌

```python
titanic.columns
```

```python
display(titanic['pclass'].value_counts())
display(titanic['pclass'].value_counts().values)  # 이런식으로 추출된 값을 배열형태로 사용할 수 있음
```

```python
# 새로운 열을 추가하고 싶다.
# age_0 : 일괄적으로 0의 값을 할당하고 싶다.

titanic['age_0'] = 0
display(titanic.info())
display(titanic.head())
```

```python
# age의 각 값에 10을 곱한 age_by_10 컬럼 생성
titanic['age_by_10'] = titanic['age'] * 10
display(titanic.head())
```

```python
# parch와 sibSp의 값과 1을 더한 family_no 컬럼생성
titanic['family_no'] = titanic['parch'] + titanic['sibsp'] + 1
```

```python
# age_by_10 칼럼 값에 일괄적으로 +100 처리
titanic['age_by_10'] = titanic['age_by_10'] + 100
```

```python
# age_0 열을 삭제하고자 한다면?
titanic.drop('age_0', axis=1).head()
```

```python
# [데이터프레임].drop('칼럼명', axis=1, inplace=True)
# inplace=True라는 옵션은 원본을 두고 복사본에 열을 삭제하는 것이 아니라
# 원본에 바로 삭제를 적용한다는 옵션이디
display(titanic.head())
titanic.drop(['age_0', 'age_by_10', 'family_no'], axis=1, inplace=True)
titanic.head()
```

```python
display(titanic.head())
# 인덱스가 0, 1, 2인 행을 삭제해보자
titanic.drop([0, 1, 2], axis=0, inplace=True)
display(titanic.head())
```

```python
# 데이터의 타입은 항상 확인하는 것이 좋다
print(type(titanic.index))
print(type(titanic.index.values))
```

```python
# 0, 1, 2를 삭제한 데이터프레임에서 상위 5개 행을 출력하라
display(titanic.iloc[0:5, ])
display(titanic.loc[titanic.index.values[0:5], ])

# index이름이 6인 행을 출력하고 싶다면?
display(titanic.loc[[6], ])
```

```python
series_fair = titanic['fare']
print('series \n', series_fair, sep='')
print()
print('type : ', type(series_fair))
```

```python
# series_fair의 max, min, sum을 가져오고 싶어
print('max :', series_fair.max())
print('min :', series_fair.min())
print('sum :', series_fair.sum())
print('=' * 50)
print('discount 10% :', series_fair * 0.9)
```
