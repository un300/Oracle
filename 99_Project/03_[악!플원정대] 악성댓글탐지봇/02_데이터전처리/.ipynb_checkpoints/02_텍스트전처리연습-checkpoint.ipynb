```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
raw_data = pd.read_csv('./내가라벨링한데이터/division_data6(재원).csv')
```

```python
df = raw_data.copy()
```

```python
df.drop(['출처', 'Unnamed: 0'], axis=1, inplace=True)
```

```python
na_index = df['악플여부'].isnull()


df = df[~na_index]
```

```python
df['댓글길이'] = df['댓글'].apply(len)
```

```python
df['음절'] = df['댓글'].apply(list)
```

```python
df.head()
```

 - 숫자 : 0-9 > 10개
 - 영어 : a-z / A-Z :

```python
syllable_list = df['음절'].tolist()
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts(syllable_list)
```

```python
tokenizer.word_index
```

```python
tokenizer.word_counts
```

```python
temp = tokenizer.texts_to_sequences(df['음절'])
```

```python
pad_sequences(temp, padding='post', maxlen=400)
```
